
# stem(x, y)

See `~matplotlib.axes.Axes.stem`.


```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make data
x = 0.5 + np.arange(8)
y = [4.8, 5.5, 3.5, 4.6, 6.5, 6.6, 2.6, 3.0]

# plot
fig, ax = plt.subplots()

ax.stem(x, y)

ax.set(xlim=(0, 8), xticks=np.arange(1, 8),
       ylim=(0, 8), yticks=np.arange(1, 8))

plt.show()
```

```python
# 매출 데이터 (단위: 억 원)
years = np.arange(2015, 2023)
revenue = [35, 40, 42, 48, 55, 60, 65, 70]

plt.style.use('_mpl-gallery')

# plot
fig, ax = plt.subplots()
fig.set_size_inches(10, 6)
ax.stem(years, revenue, linefmt="C0-", markerfmt="C0o", basefmt=" ")

ax.set(xlabel='연도', ylabel='매출액 (억 원)',
       title='연간 매출 추이')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 매출 데이터 (단위: 억 원)
years = np.arange(2015, 2023)
revenue = [35, 40, 42, 48, 55, 60, 65, 70]

# 영업 이익 데이터 (단위: 억 원)
profit = [10, 12, 15, 20, 22, 25, 28, 30]

plt.style.use('_mpl-gallery')

# plot
fig, (ax1, ax2) = plt.subplots(1, 2)  # 1행 2열로 subplot 생성
fig.set_size_inches(16, 6)  # 그림 크기 조정

# 첫 번째 subplot: 매출
ax1.stem(years, revenue, linefmt="C0-", markerfmt="C0o", basefmt=" ")
ax1.set(xlabel='연도', ylabel='매출액 (억 원)', title='연간 매출 추이')

# 두 번째 subplot: 이익
ax2.stem(years, profit, linefmt="C1-", markerfmt="C1o", basefmt=" ")
ax2.set(xlabel='연도', ylabel='이익 (억 원)', title='연간 영업 이익 추이')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 매출 데이터 (단위: 억 원)
years = np.arange(2015, 2023)
revenue = [35, 40, 42, 48, 55, 60, 65, 70]

# 영업 이익 데이터 (단위: 억 원)
profit = [10, 12, 15, 20, 22, 25, 28, 30]

plt.style.use('_mpl-gallery')

# plot
fig, (ax1, ax2) = plt.subplots(1, 2)  # 1행 2열로 subplot 생성
fig.set_size_inches(16, 6)  # 그림 크기 조정

# 첫 번째 subplot: 매출
ax1.stem(years, revenue, linefmt="C0-", markerfmt="C0o", basefmt=" ", label='매출')
ax1.set(xlabel='연도', ylabel='매출액 (억 원)', title='연간 매출 추이')
ax1.legend()

# 두 번째 subplot: 이익
ax2.stem(years, profit, linefmt="C1-", markerfmt="C1o", basefmt=" ", label='이익')
ax2.set(xlabel='연도', ylabel='이익 (억 원)', title='연간 영업 이익 추이')
ax2.legend()

plt.show()
```
