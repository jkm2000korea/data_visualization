
# Scatter Masked

Mask some data points and add a line demarking
masked regions.


```python
import matplotlib.pyplot as plt
import numpy as np

# 재현성을 위해 랜덤 시드 설정
np.random.seed(19680801)

N = 100  # 데이터 포인트의 개수
r0 = 0.6  # 중심 반경

x = 0.9 * np.random.rand(N)  # x 좌표 생성
y = 0.9 * np.random.rand(N)  # y 좌표 생성

area = (20 * np.random.rand(N))**2  # 0에서 10 사이의 점 반지름 생성
c = np.sqrt(area)  # 색상값 생성

r = np.sqrt(x ** 2 + y ** 2)  # 중심에서의 거리 계산
area1 = np.ma.masked_where(r < r0, area)  # 중심 내부의 영역
area2 = np.ma.masked_where(r >= r0, area)  # 중심 외부의 영역
plt.figure(figsize=(7, 7))
plt.scatter(x, y, s=area1, marker='^', c=c)  # 중심 내부의 점들 플롯
plt.scatter(x, y, s=area2, marker='o', c=c)  # 중심 외부의 점들 플롯


# 영역 간 경계 표시:
theta = np.arange(0, np.pi / 2, 0.01)  # 0부터 pi/2까지의 각도 생성
plt.plot(r0 * np.cos(theta), r0 * np.sin(theta))  # 중심 경계선 플롯

plt.show()  # 그래프 표시
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 재현성을 위해 랜덤 시드 설정
np.random.seed(0)

r_values = [40, 70, 100]  # 각 중심의 반경

# 데이터 포인트 생성
N = 40

# 40, 40 미만
x1 = np.random.randint(0, 35, 10)
y1 = np.random.randint(0, 35, 10)

# (40, 40)과 (60, 60) 사이
x2 = np.random.randint(35, 50, 10)
y2 = np.random.randint(35, 50, 10)

# (60, 60)과 (80, 80) 사이
x3 = np.random.randint(50, 75, 10)
y3 = np.random.randint(50, 75, 10)


# 중심 내부의 점들 플롯
plt.figure(figsize=(7, 7))

plt.scatter(x1, y1, marker='^', color='royalblue')
plt.scatter(x2, y2, marker='o', color='blue')
plt.scatter(x3, y3, marker='*', color='lightblue') 

# 중심 경계선 플롯
theta = np.arange(0, np.pi / 2, 0.01)  # 0부터 pi/2까지의 각도 생성
plt.plot(r_values[0] * np.cos(theta), r_values[0] * np.sin(theta), color='royalblue')  # 첫 번째 중심 경계선 플롯
plt.plot(r_values[1] * np.cos(theta), r_values[1] * np.sin(theta), color='blue')  # 두 번째 중심 경계선 플롯
plt.plot(r_values[2] * np.cos(theta), r_values[2] * np.sin(theta), color='lightblue')  # 세 번째 중심 경계선 플롯

# 각 영역을 음영으로 채우기
theta = np.linspace(0, np.pi/2, 100)
plt.fill_between(np.arange(0, r_values[0] + 1), 0, np.sqrt(r_values[0] ** 2 - np.arange(0, r_values[0] + 1) ** 2), color='lightblue', alpha=0.3)
plt.fill_between(np.arange(0, r_values[1] + 1), 0, np.sqrt(r_values[1] ** 2 - np.arange(0, r_values[1] + 1) ** 2), color='blue', alpha=0.2)
plt.fill_between(np.arange(0, r_values[2] + 1), 0, np.sqrt(r_values[2] ** 2 - np.arange(0, r_values[2] + 1) ** 2), color='royalblue', alpha=0.1)

# 영역별 우선순위 텍스트 표시
plt.text(r_values[0] - 15, 10, 'I', fontsize=20, color='black', fontweight='bold')
plt.text(r_values[1] - 15, 10, 'II', fontsize=20, color='black', fontweight='bold')
plt.text(r_values[2] - 15, 10, 'III', fontsize=20, color='black', fontweight='bold')
plt.margins(x=0, y=0)
plt.title("McKinsey Priority Matrix",  fontsize=20)
plt.show()  # 그래프 표시
```
